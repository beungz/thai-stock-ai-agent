```python
# Initialize the Scraper
scraper = SETStockData(version=146)

# Wrap it in the Stateful Tools class (Handles caching)
tool_manager = SETAgentTools(scraper)
    
# Initialize the Agent with the desired model and the tool manager
agent = LocalFinancialAgent(
    model_name="qwen2.5:14b-instruct", 
    tool_manager=tool_manager, 
    registry_path="function_registry.json"
)
```

```python
# Send a complex query
prompt = "I want to compare the profitability of PTT against its peers in the same subsector for the year 2023. Please get the data and show me the table."
agent.run(prompt)
```

```python
prompt = "I want to see the profitability of PTT against BCP, SPRC, PTTGC, and IRPC in 2025. Please get the data and show me the table."
agent.run(prompt)
```
